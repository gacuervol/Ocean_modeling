# Fuerza de coriolis
---

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Periodo de rotacion del tanque
T = 24 * 60 * 60 # s (24 h)
# Radio del tanque de agua
r = 10_000 # m
# Velocidad de rotacion
Omega = (2 * np.pi)/T # rad/s 
# Fuerza centripeta
F_c = - Omega**2 * r # 
```

```python
dt = 60 # s
n_iter = 5000
# Condiciones iniciales
X_0, Y_0 = 0, 5_000 # m 
U_0, V_0 =  0.864, 0.5 # m/s 
# Diferencias finitas
X_arr = np.array([], dtype=np.float64)
Y_arr = np.array([], dtype=np.float64)
while (X_arr.size == Y_arr.size) and (X_arr.size < n_iter):
    U_n, V_n = U_0 - dt * Omega**2 * X_0, V_0 - dt * Omega**2 * Y_0  
    X_n, Y_n = X_0 + dt * U_n, Y_0 + dt * V_n 
    X_n, Y_n = np.array(X_n, dtype=np.float64), np.array(Y_n, dtype=np.float64)
    X_arr, Y_arr = np.concatenate((X_arr, [X_n])), np.concatenate((Y_arr, [Y_n]))
    U_0, V_0 = U_n, V_n
    X_0, Y_0 = X_n, Y_n
```

```python
plt.plot(X_arr, Y_arr);
```
